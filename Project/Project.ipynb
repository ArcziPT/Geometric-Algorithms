```python
def det3(a, b, c):
    return (a[0]*b[1] + b[0]*c[1] + c[0]*a[1]) - (c[0]*b[1] + b[0]*a[1] + a[0]*c[1])

class Edge:
    def __init__(self, p1, p2):
        self.p1 = p1 #first point
        self.p2 = p2 #second point
        
    def __eq__(self, other):
        #symetric edges
        return (self.p1 == other.p1 and self.p2 == other.p2) or (self.p1 == other.p2 and self.p2 == other.p1)
    
    def __hash__(self):
        return hash((self.p1, self.p2))
    
    def __str__(self):
        return "{" + str(self.p1) + "," + str(self.p2) + "}"
    
    def __repr__(self):
        return "{" + str(self.p1) + "," + str(self.p2) + "}"
    
class Tr:
    def __init__(self, e1, e2, e3):
        #e1,e2,e3 - edges of triangle CCW
        self.edges = set([e1, e2, e3])
        
    def circumcircle(self):
        '''
        midpoint and radius of circumcircle
        https://en.wikipedia.org/wiki/Circumscribed_circle
        '''
        pts = []
        for edge in self.edges:
            pts.append(edge.p1)
        
        a = pts[0][0]**2 + pts[0][1]**2
        b = pts[1][0]**2 + pts[1][1]**2
        c = pts[2][0]**2 + pts[2][1]**2
        
        t = det3(pts[0], pts[1], pts[2])
        
        sx = (1/2) * det3([a, pts[0][1]],[b, pts[1][1]], [c, pts[2][1]]) / t
        sy = (1/2) * det3([pts[0][0], a],[pts[1][0], b], [pts[2][0], c]) / t
        s = np.array([sx, sy])
        
        return (s, np.linalg.norm(s-pts[0]))
```

```python
def triangulate(points):
    #random shuffle
    points = np.random.permutation(points)
    
    y_max, y_min = max(points[:,1].tolist()), min(points[:,1].tolist())
    x_max, x_min = max(points[:,0].tolist()), min(points[:,0].tolist())
    
    m = max([abs(y_max), abs(y_min), abs(x_max), abs(x_min)])
    
    #create super-triangle
    superTr = Tr(Edge((-10*m, -10*m), (0, 10*m)), Edge((0, 10*m), (10*m, 0)), Edge((10*m, 0), (-10*m, -10*m)))
    triangulation = [superTr]
    
    #triangulation
    #https://en.wikipedia.org/wiki/Bowyer%E2%80%93Watson_algorithm
    for p in points:
        badTrs = [] #triangles to be removed (whose cicumcirlce contains new point)
        goodTrs = [] #others
        for tr in triangulation:
            s, r = tr.circumcircle()
            if np.linalg.norm(p-s) <= r:
                badTrs.append(tr)
            else:
                goodTrs.append(tr)
                
        edges = {} #number of edge occurrences in badTrs 
        for tr in badTrs:
            for e in tr.edges:
                if e in edges:
                    edges[e] += 1
                else:
                    if Edge(e.p2, e.p1) in edges: #symetric edge
                        edges[Edge(e.p2, e.p1)] += 1
                    else:
                        edges[e] = 1
        
        triangulation = goodTrs #remove badTrs
        
        for e in edges:
            #if edge occured only once it is the edge of polygon, which contains all of the remove triangles
            if edges[e] == 1:
                #add new triangle
                triangulation.append(Tr(e, Edge(e.p2, tuple(p.tolist())), Edge(tuple(p.tolist()), e.p1)))    
    
    #remove edges to super-triangle
    edges = set()
    for tr in triangulation:
        for e in tr.edges:
            toAdd = True
            for sE in superTr.edges:
                if e.p1 == sE.p1 or e.p2 == sE.p1 or e.p1 == sE.p2 or e.p2 == sE.p2:
                    toAdd = False
            if toAdd:
                edges.add(e)
    
    return edges
```

```python
%matplotlib notebook
def plot_res(res):
    for e in res:
        plt.plot([e.p1[0], e.p2[0]], [e.p1[1], e.p2[1]])
```

```python
pts = np.random.uniform(-1000, 1000, size=(20, 2))
res = triangulate(pts)
plot_res(res)
```

# Animation

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections as mcoll
import matplotlib.colors as mcolors
from matplotlib.widgets import Button
import json as js
import timeit
import time
```

```python
# Parametr okreÅ›lajÄ…cy jak blisko (w odsetku caÅ‚ego widocznego zakresu) punktu poczÄ…tkowego 
# wielokÄ…ta musimy kliknÄ…Ä‡, aby go zamknÄ…Ä‡.
TOLERANCE = 0.15

def dist(point1, point2):
    return np.sqrt(np.power(point1[0] - point2[0], 2) + np.power(point1[1] - point2[1], 2))

# Klasa ta trzyma obecny stan wykresu oraz posiada metody, ktÃ³re majÄ… zostaÄ‡ wykonane
# po naciÅ›niÄ™ciu przyciskÃ³w.
class _Button_callback(object):
    def __init__(self, scenes, x_lim = None, y_lim = None):
        self.i = 0
        self.scenes = scenes
        self.adding_points = False
        self.added_points = []
        self.adding_lines = False
        self.added_lines = []
        self.adding_rects = False
        self.added_rects = []
        self.x_lim = x_lim
        self.y_lim = y_lim

    def set_axes(self, ax):
        self.ax = ax
        
    # Metoda ta obsÅ‚uguje logikÄ™ przejÅ›cia do nastÄ™pnej sceny.
    def next(self, event):
        self.i = (self.i + 1) % len(self.scenes)
        self.draw(autoscaling = True)

    # Metoda ta obsÅ‚uguje logikÄ™ powrotu do poprzedniej sceny.
    def prev(self, event):
        self.i = (self.i - 1) % len(self.scenes)
        self.draw(autoscaling = True)
        
    # Metoda ta aktywuje funkcjÄ™ rysowania punktÃ³w wyÅ‚Ä…czajÄ…c rÃ³wnoczeÅ›nie rysowanie 
    # odcinkÃ³w i wielokÄ…tÃ³w.
    def add_point(self, event):
        self.adding_points = not self.adding_points
        self.new_line_point = None
        if self.adding_points:
            self.adding_lines = False
            self.adding_rects = False
            self.added_points.append(PointsCollection([]))
            
    # Metoda ta aktywuje funkcjÄ™ rysowania odcinkÃ³w wyÅ‚Ä…czajÄ…c rÃ³wnoczeÅ›nie
    # rysowanie punktÃ³w i wielokÄ…tÃ³w.     
    def add_line(self, event):   
        self.adding_lines = not self.adding_lines
        self.new_line_point = None
        if self.adding_lines:
            self.adding_points = False
            self.adding_rects = False
            self.added_lines.append(LinesCollection([]))

    # Metoda ta aktywuje funkcjÄ™ rysowania wielokÄ…tÃ³w wyÅ‚Ä…czajÄ…c rÃ³wnoczeÅ›nie
    # rysowanie punktÃ³w i odcinkÃ³w.
    def add_rect(self, event):
        self.adding_rects = not self.adding_rects
        self.new_line_point = None
        if self.adding_rects:
            self.adding_points = False
            self.adding_lines = False
            self.new_rect()
    
    def new_rect(self):
        self.added_rects.append(LinesCollection([]))
        self.rect_points = []
        
    # Metoda odpowiedzialna za wÅ‚aÅ›ciwÄ… logikÄ™ rysowania nowych elementÃ³w. W
    # zaleÅ¼noÅ›ci od wÅ‚Ä…czonego trybu dodaje nowe punkty, poczÄ…tek, koniec odcinka
    # lub poszczegÃ³lne wierzchoÅ‚ki wielokÄ…ta. Istnieje ciekawa logika sprawdzania
    # czy dany punkt jest domykajÄ…cy dla danego wielokÄ…ta. Polega ona na tym, Å¼e
    # sprawdzamy czy odlegÅ‚oÅ›Ä‡ nowego punktu od poczÄ…tkowego jest wiÄ™ksza od
    # Å›redniej dÅ‚ugoÅ›ci zakresu pomnoÅ¼onej razy parametr TOLERANCE.   
    def on_click(self, event):
        if event.inaxes != self.ax:
            return
        new_point = (event.xdata, event.ydata)
        if self.adding_points:
            self.added_points[-1].add_points([new_point])
            self.draw(autoscaling = False)
        elif self.adding_lines:
            if self.new_line_point is not None:
                self.added_lines[-1].add([self.new_line_point, new_point])
                self.new_line_point = None
                self.draw(autoscaling = False)
            else:
                self.new_line_point = new_point
        elif self.adding_rects:
            if len(self.rect_points) == 0:
                self.rect_points.append(new_point)
            elif len(self.rect_points) == 1:
                self.added_rects[-1].add([self.rect_points[-1], new_point])
                self.rect_points.append(new_point)
                self.draw(autoscaling = False)
            elif len(self.rect_points) > 1:
                if dist(self.rect_points[0], new_point) < (np.mean([self.ax.get_xlim(), self.ax.get_ylim()])*TOLERANCE):
                    self.added_rects[-1].add([self.rect_points[-1], self.rect_points[0]])
                    self.new_rect()
                else:    
                    self.added_rects[-1].add([self.rect_points[-1], new_point])
                    self.rect_points.append(new_point)
                self.draw(autoscaling = False)
    
    # Metoda odpowiedzialna za narysowanie caÅ‚ego wykresu. Warto zauwaÅ¼yÄ‡,
    # Å¼e zaczyna siÄ™ ona od wyczyszczenia jego wczeÅ›niejszego stanu. Istnieje w
    # niej nietrywialna logika zarzÄ…dzania zakresem wykresu, tak Å¼eby, w zaleÅ¼noÅ›ci
    # od ustawionego parametru autoscaling, uniknÄ…Ä‡ sytuacji, kiedy dodawanie
    # nowych punktÃ³w przy brzegu obecnie widzianego zakresu powoduje niekorzystne
    # przeskalowanie.
    def draw(self, autoscaling = True):
        if not autoscaling:
            xlim = self.ax.get_xlim()
            ylim = self.ax.get_ylim()
        self.ax.clear()
        for collection in (self.scenes[self.i].points + self.added_points):
            if len(collection.points) > 0:
                self.ax.scatter(*zip(*(np.array(collection.points))), **collection.kwargs)
        for collection in (self.scenes[self.i].lines + self.added_lines + self.added_rects):
            self.ax.add_collection(collection.get_collection())
        if self.x_lim is not None and self.y_lim is not None:
            self.ax.set_xlim(self.x_lim)
            self.ax.set_ylim(self.y_lim)
        else:
            self.ax.autoscale(autoscaling)
        
        if not autoscaling:
            self.ax.set_xlim(xlim)
            self.ax.set_ylim(ylim)
        plt.draw()
```

```python
# Klasa Scene odpowiada za przechowywanie elementÃ³w, ktÃ³re majÄ… byÄ‡
# wyÅ›wietlane rÃ³wnoczeÅ›nie. Konkretnie jest to lista PointsCollection i
# LinesCollection.
class Scene:
    def __init__(self, points=[], lines=[]):
        self.points=points
        self.lines=lines

# Klasa PointsCollection gromadzi w sobie punkty jednego typu, a wiÄ™c takie,
# ktÃ³re zostanÄ… narysowane w takim samym kolorze i stylu. W konstruktorze
# przyjmuje listÄ™ punktÃ³w rozumianych jako pary wspÃ³Å‚rzÄ™dnych (x, y). Parametr
# kwargs jest przekazywany do wywoÅ‚ania funkcji z biblioteki MatPlotLib przez
# co uÅ¼ytkownik moÅ¼e podawaÄ‡ wszystkie parametry tam zaproponowane.        
class PointsCollection:
    def __init__(self, points, **kwargs):
        self.points = points
        self.kwargs = kwargs
    
    def add_points(self, points):
        self.points = self.points + points

# Klasa LinesCollection podobnie jak jej punktowy odpowiednik gromadzi
# odcinki tego samego typu. TworzÄ…c jÄ… naleÅ¼y podaÄ‡ listÄ™ linii, gdzie kaÅ¼da
# z nich jest dwuelementowÄ… listÄ… punktÃ³w â€“ par (x, y). Parametr kwargs jest
# przekazywany do wywoÅ‚ania funkcji z biblioteki MatPlotLib przez co uÅ¼ytkownik
# moÅ¼e podawaÄ‡ wszystkie parametry tam zaproponowane.
class LinesCollection:
    def __init__(self, lines, **kwargs):
        self.lines = lines
        self.kwargs = kwargs
        
    def add(self, line):
        self.lines.append(line)
        
    def get_collection(self):
        return mcoll.LineCollection(self.lines, **self.kwargs)

# Klasa Plot jest najwaÅ¼niejszÄ… klasÄ… w caÅ‚ym programie, poniewaÅ¼ agreguje
# wszystkie przygotowane sceny, odpowiada za stworzenie wykresu i przechowuje
# referencje na przyciski, dziÄ™ki czemu nie bÄ™dÄ… one skasowane podczas tzw.
# garbage collectingu.
class Plot:
    def __init__(self, scenes = [Scene()], points = [], lines = [], x_lim = None, y_lim = None, json = None):
        self.x_lim = x_lim
        self.y_lim = y_lim
        
        if json is None:
            self.scenes = scenes
            if points or lines:
                self.scenes[0].points = points
                self.scenes[0].lines = lines
        else:
            self.scenes = [Scene([PointsCollection(pointsCol) for pointsCol in scene["points"]], 
                                 [LinesCollection(linesCol) for linesCol in scene["lines"]]) 
                           for scene in js.loads(json)]
    
    # Ta metoda ma szczegÃ³lne znaczenie, poniewaÅ¼ konfiguruje przyciski i
    # wykonuje tym samym doÅ›Ä‡ skomplikowanÄ… logikÄ™. ZauwaÅ¼my, Å¼e konfigurujÄ…c kaÅ¼dy
    # przycisk podajemy referencjÄ™ na metodÄ™ obiektu _Button_callback, ktÃ³ra
    # zostanie wykonana w momencie naciÅ›niÄ™cia.
    def __configure_buttons(self):
        plt.subplots_adjust(bottom=0.2)
        ax_prev = plt.axes([0.6, 0.05, 0.15, 0.075])
        ax_next = plt.axes([0.76, 0.05, 0.15, 0.075])
        ax_add_point = plt.axes([0.44, 0.05, 0.15, 0.075])
        ax_add_line = plt.axes([0.28, 0.05, 0.15, 0.075])
        ax_add_rect = plt.axes([0.12, 0.05, 0.15, 0.075])
        b_next = Button(ax_next, 'NastÄ™pny')
        b_next.on_clicked(self.callback.next)
        b_prev = Button(ax_prev, 'Poprzedni')
        b_prev.on_clicked(self.callback.prev)
        b_add_point = Button(ax_add_point, 'Dodaj punkt')
        b_add_point.on_clicked(self.callback.add_point)
        b_add_line = Button(ax_add_line, 'Dodaj liniÄ™')
        b_add_line.on_clicked(self.callback.add_line)
        b_add_rect = Button(ax_add_rect, 'Dodaj figurÄ™')
        b_add_rect.on_clicked(self.callback.add_rect)
        return [b_prev, b_next, b_add_point, b_add_line, b_add_rect]
    
    def add_scene(self, scene):
        self.scenes.append(scene)
    
    def add_scenes(self, scenes):
        self.scenes = self.scenes + scenes

    # Metoda toJson() odpowiada za zapisanie stanu obiektu do ciÄ…gu znakÃ³w w
    # formacie JSON.
    def toJson(self):
        return js.dumps([{"points": [np.array(pointCol.points).tolist() for pointCol in scene.points], 
                          "lines":[linesCol.lines for linesCol in scene.lines]} 
                         for scene in self.scenes])    
    
    # Metoda ta zwraca punkty dodane w trakcie rysowania.
    def get_added_points(self):
        if self.callback:
            return self.callback.added_points
        else:
            return None
    
    # Metoda ta zwraca odcinki dodane w trakcie rysowania.
    def get_added_lines(self):
        if self.callback:
            return self.callback.added_lines
        else:
            return None
        
    # Metoda ta zwraca wielokÄ…ty dodane w trakcie rysowania.
    def get_added_figure(self):
        if self.callback:
            return self.callback.added_rects
        else:
            return None
    
    # Metoda ta zwraca punkty, odcinki i wielokÄ…ty dodane w trakcie rysowania
    # jako scenÄ™.
    def get_added_elements(self):
        if self.callback:
            return Scene(self.callback.added_points, self.callback.added_lines+self.callback.added_rects)
        else:
            return None
    
    # GÅ‚Ã³wna metoda inicjalizujÄ…ca wyÅ›wietlanie wykresu.
    def draw(self):
        plt.close()
        fig = plt.figure()
        self.callback = _Button_callback(self.scenes, self.x_lim, self.y_lim)
        self.widgets = self.__configure_buttons()
        ax = plt.axes(autoscale_on = False)
        self.callback.set_axes(ax)
        fig.canvas.mpl_connect('button_press_event', self.callback.on_click)
        plt.show()
        self.callback.draw()
```

```python
def get_lines(trs, superTr):
    edges = set()
    for tr in trs:
        for e in tr.edges:
            toAdd = True
            for sE in superTr.edges:
                if e.p1 == sE.p1 or e.p2 == sE.p1 or e.p1 == sE.p2 or e.p2 == sE.p2:
                    toAdd = False
            if toAdd:
                edges.add(e)
    ret = []
    for e in edges:
        ret.append([e.p1, e.p2])
    return ret

def triangulate_scenes(points):
    scenes = []
    
    points = np.random.permutation(points)
    
    y_max, y_min = max(points[:,1].tolist()), min(points[:,1].tolist())
    x_max, x_min = max(points[:,0].tolist()), min(points[:,0].tolist())
    
    m = max([abs(y_max), abs(y_min), abs(x_max), abs(x_min)])
     
    superTr = Tr(Edge((-10*m, -10*m), (0, 10*m)), Edge((0, 10*m), (10*m, 0)), Edge((10*m, 0), (-10*m, -10*m)))
    triangulation = [superTr]
    #triangulation = [Tr(Edge((-1000, -1000), (1000, -1000)), Edge((1000, -1000), (0, 1000)), Edge((0, 1000), (-1000, -1000)))]
    for p in points:
        badTrs = [] #do usuniÄ™cia
        goodTrs = [] #pozostaÅ‚e
        for tr in triangulation:
            s, r = tr.circumcircle()
            if np.linalg.norm(p-s) <= r:
                badTrs.append(tr)
            else:
                goodTrs.append(tr)
                
        edges = {} #liczba wystÄ™pieÅ„ krawÄ™dzi
        for tr in badTrs:
            for e in tr.edges:
                if e in edges:
                    edges[e] += 1
                else:
                    if Edge(e.p2, e.p1) in edges:
                        edges[Edge(e.p2, e.p1)] += 1
                    else:
                        edges[e] = 1
        
        triangulation = goodTrs
        
        oldEdges = get_lines(triangulation, superTr)
        scenes.append(Scene(lines=[LinesCollection(oldEdges.copy(), color="blue")]))
        
        newTr = []
        for e in edges:
            # jeÅ¼eli krawÄ™dÅº wystÄ…piÅ‚a tylko raz to jest to krawÄ™dÅº poszukiwanego wielokÄ…ta
            if edges[e] == 1:
                triangulation.append(Tr(e, Edge(e.p2, tuple(p.tolist())), Edge(tuple(p.tolist()), e.p1)))
                newTr.append(Tr(e, Edge(e.p2, tuple(p.tolist())), Edge(tuple(p.tolist()), e.p1)))
                
        scenes.append(Scene(lines=[LinesCollection(oldEdges.copy(), color="blue"), LinesCollection(get_lines(newTr, superTr), color="red")]))
        scenes.append(Scene(lines=[LinesCollection(get_lines(triangulation, superTr), color="blue")]))
    
    return scenes, get_lines(triangulation, superTr)
```

```python
def animate(points):
    scenes, res = triangulate_scenes(points)
    
    x_max, x_min = max(points[:,0]), min(points[:,0])
    y_max, y_min = max(points[:,1]), min(points[:,1])
    
    plot = Plot(scenes, x_lim=(1.1*x_min, 1.1*x_max), y_lim=(1.1*y_min, 1.1*y_max))
    plot.draw()
    return plot
```
